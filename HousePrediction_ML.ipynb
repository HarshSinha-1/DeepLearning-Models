**UNIVARIATE ANALYSIS**

```python
print("\nUnivariate Analysis:")
plt.figure(figsize=(10, 6))
sns.histplot(data['Y house price of unit area'], bins=30, kde=True)
plt.title("House Price Distribution")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.show()
```

The distribution of house prices is right-skewed, as indicated by the histogram and the skewness value of approximately 2.83, which is significantly greater than zero. This suggests that the data has a long tail on the higher end of the price range.
